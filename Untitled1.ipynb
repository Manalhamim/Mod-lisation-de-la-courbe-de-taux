```python
dg=df.copy()
a=np.max(dg['Tmp'])
b=np.max(dg['Tmp'])-np.min(dg['Tmp'])
c=2*np.mean(dg['Tmp'])-np.max(dg['Tmp'])-np.min(dg['Tmp'])
d=c
vj=[a,b,c,d]
def SPC(m,vj):
    f=vj[0]+vj[1]*m+vj[2]*m**2+vj[3]*m**3
    return f
dg['SPC'] = SPC(dg['ttm'], vj)
```

```python
def VasicekCalibration(r, dt=1/252):
    n = len(r)
    # Implement MLE to calibrate parameters     
    Sx = np.sum(r[0:(n-1)])
    Sy = np.sum(r[1:n])
    Sxx = np.dot(r[0:(n-1)], r[0:(n-1)])
    Sxy = np.dot(r[0:(n-1)], r[1:n])
    Syy = np.dot(r[1:n], r[1:n])
    a= (Sy * Sxx - Sx * Sxy) / (n * (Sxx - Sxy) - (Sx**2 - Sx*Sy))
    b= -np.log((Sxy - a* Sx - a* Sy + n * a**2) / (Sxx - 2*a*Sx + n*a**2)) / dt
    c= np.exp(-b * dt)
    sigmah2 = (Syy - 2*c*Sxy + c**2 * Sxx - 2*a*(1-c)*(Sy - c*Sx) + n*a**2 * (1-c)**2) / n
    sigma = np.sqrt(sigmah2*2*b / (1-c**2))
    r0 = r[0]
    return [a, b, sigma, r0]
```

```python
def vasicek(r0, a, b, sigma, T, N):
    dt = T/N
    r = np.zeros(N+1)
    r[0] = r0
    for i in range(1, N+1):
        e = np.random.normal(0, 1)
        r[i] = r[i-1] + a*(b-r[i-1])*dt + sigma*np.sqrt(dt)*e
    return r
```

```python
def CIR(r2, dt=1/252):
    n = len(r2)
    # Implement MLE to calibrate parameters     
    Sx = np.sum(r2[0:(n-1)])
    Sy = np.sum(r2[1:n])
    Sxx = np.dot(r2[0:(n-1)], r2[0:(n-1)])
    Sxy = np.dot(r2[0:(n-1)], r2[1:n])
    Syy = np.dot(r2[1:n], r2[1:n])
    b= (Sy * Sxx - Sx * Sxy) / (n * (Sxx - Sxy) - (Sx**2 - Sx*Sy))
    a = -np.log((Sxy - b * Sx - b * Sy + n * b**2) / (Sxx - 2*b*Sx + n*b**2)) / dt
    c= np.exp(-a * dt)
    sigmah2 = (Syy - 2*c*Sxy + c**2 * Sxx - 2*b*(1-c)*(Sy - c*Sx) + n*b**2 * (1-c)**2) / n
    sigma = np.sqrt(sigmah2*2*a/ (1-c**2))
    r0 = r2[n-1]
    return [a, c, sigma, r0]
```
